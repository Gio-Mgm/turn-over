```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv("../data/01_raw/attrition_train.csv")

```

```python
cat_cols = train.select_dtypes(include="O").columns
cat_cols = cat_cols.drop(["Attrition"])
for col in train[cat_cols]:
    plt.figure(figsize=(10,4))
    sns.countplot(x=train[col], hue=train["Attrition"])
```

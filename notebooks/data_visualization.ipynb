```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv("../data/01_raw/attrition_train.csv")

```

```python
cat_cols = train.select_dtypes(include="O").columns

cat_cols = cat_cols.drop(["Attrition", "Over18", "EducationField", "JobRole"])
for col in train[cat_cols]:
    sns.displot(data=train, x=col, hue='Attrition',  stat="density", common_norm=False, multiple="dodge", height=3, aspect=1.5, shrink=0.9)
```

```python
num_cols = train.select_dtypes(exclude="O").columns

num_cols = num_cols.drop(["EmployeeCount"])
for col in train[num_cols]:
    sns.displot(data=train, x=col, hue='Attrition',  stat="density", common_norm=False, multiple="dodge", height=5, aspect=1.5, shrink=0.9, kde=True)
```
